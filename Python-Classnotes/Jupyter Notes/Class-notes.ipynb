```python
year = int(input("Enter year to check"))
if year % 400 == 0:
    print(year, "is a century Lea[ Year!")
elif year % 100 == 0:
    print(year, "is not a leap year(century year).")
elif year % 4 == 0:
    print(year, "is a leap year!")
else:
    print(year, "is a common year.")
```

```python
# 1-200 , 2.70 paisa per unit
# 201-400 , 4 rs per unit
# 401-700 7 per unit
# >700 9 per unit

units = float(input("Enter total units consumed: "))
bill = 0

if units <= 0:
    print("Wrong Input, please enter a positive value.")
elif units <= 200:
    # 1-200 range
    bill = units * 2.70
elif units <= 400:
    # 201-400 range
    units = units - 200  # Remove the first 200 units
    bill = (units * 4) + 540  # 540 is the cost of the first 200 units (200 * 2.7)
elif units <= 700:
    # 401-700 range
    units = units - 400  # Remove the first 400 units
    bill = (units * 7) + 1340  # 1340 is the cost of first 400 units (540 + 800)
else:
    # > 700 range
    units = units - 700  # Remove the first 700 units
    bill = (units * 9) + 3440  # 3440 is the cost of first 700 units (1340 + 2100)

if units > 0:
    print("Total Electricity Bill = Rs.", bill)
```

```python
# Matchstick Game
import random

ms = 21

while ms > 0:
    # --- Player 1 ki Turn ---
    p1 = random.randrange(1, 4)
    print("Matchstick selected by Player 1 =", p1)
    ms = ms - p1

    if ms <= 0:
        print("Matchsticks remaining = 0")
        print("Player 1 picked the last stick! Player 1 LOST the match. ❌")
        break  # Yahan se loop turant khatam ho jayega

    print("Matchsticks remaining =", ms)
    print("-" * 20)

    # --- Player 2 ki Turn ---
    p2 = random.randrange(1, 4)
    print("Matchstick selected by Player 2 =", p2)
    ms = ms - p2

    if ms <= 0:
        print("Matchsticks remaining = 0")
        print("Player 2 picked the last stick! Player 2 LOST the match. ❌")
        break  # Yahan se loop turant khatam ho jayega

    print("Matchsticks remaining =", ms)
    print("-" * 20)

print("\n--- Game Over ---")
```

```python
import pymongo

client = pymongo.MongoClient("mongodb://localhost:27017/")
db = client["pyproject"]


def Menuandbill(name, phone):
    # Data ko list mein rakha hai taaki loop chal sake
    MenuItems = [
        "Pizza",
        "Burger",
        "Chowmin",
        "Idli",
        "Dosa",
        "Wada",
        "Shahi Paneer",
        "Khadai Paneer",
        "Matar Paneer",
        "Garlic Naan",
        "Butter Roti",
        "Rumali Roti",
    ]
    Pricing = [150, 50, 80, 60, 100, 250, 100, 150, 250, 20, 15, 5]

    selected_items = {}  # User ki choice store karne ke liye: {item_index: quantity}

    print(f"\nWelcome {name}! Our Menu:-\n")
    for i, item in enumerate(MenuItems):
        print(f"{i+1}) {item} :- ₹{Pricing[i]}")

    while True:
        choice = input("\nEnter Item No. (or 'q' to finish): ")
        if choice.lower() == "q":
            break

        if choice.isdigit():
            idx = int(choice) - 1
            if 0 <= idx < len(MenuItems):
                qty = int(input(f"Enter quantity for {MenuItems[idx]}: "))
                if qty <= 0:
                    qty = 1

                # Agar item pehle se hai toh quantity add hogi
                selected_items[idx] = selected_items.get(idx, 0) + qty
                print(f"{MenuItems[idx]} added.")
            else:
                print("Invalid Item Number!")
        else:
            print("Please enter a valid digit.")

        cont = input("Press 1 to add more, any other key to generate bill: ")
        if cont != "1":
            break

    # Bill Generation Logic
    total_pricing = 0
    var_all_order = []

    print("\n" + "=" * 50)
    print(f"{'Items':<20}{'Qty':<10}{'Price':<10}{'Total':<10}")
    print("-" * 50)

    for idx, qty in selected_items.items():
        item_name = MenuItems[idx]
        unit_price = Pricing[idx]
        line_total = qty * unit_price
        total_pricing += line_total

        var_all_order.append(
            {
                "Item": item_name,
                "Quantity": qty,
                "Itemprice": unit_price,
                "Itemtotal": line_total,
            }
        )
        print(f"{item_name:<20}{qty:<10}₹{unit_price:<10}₹{line_total:<10}")

    # Database Entry
    main_order = {
        "userphoneno": phone,
        "order": var_all_order,
        "Ordervalue": total_pricing,
    }
    db["orders"].insert_one(main_order)

    print("-" * 50)
    print(f"{'GRAND TOTAL':<40}₹{total_pricing}")
    print("=" * 50)


def printbill(arr, orderval):
    print("\n--- Your Last Order ---")
    print(f"{'Items':<20}{'Qty':<10}{'Price':<10}{'Total':<10}")
    for i in arr:
        print(
            f"{i['Item']:<20}{i['Quantity']:<10}₹{i['Itemprice']:<10}₹{i['Itemtotal']:<10}"
        )
    print(f"\nTotal Paid: ₹{orderval}\n")


# Main Logic
print("Welcome to our Restaurant")
print("1: New Customer\n2: Existing Customer\n3: Find Order")
main_choice = input("Enter Choice: ")

if main_choice == "1":
    name = input("Name: ")
    phone = input("Phone (10 digits): ")
    if phone.isdigit() and len(phone) == 10:
        if db["users"].find_one({"Phone": phone}):
            print("User already exists!")
        else:
            db["users"].insert_one({"Name": name, "Phone": phone})
            Menuandbill(name, phone)
elif main_choice == "2":
    phone = input("Enter Phone: ")
    user = db["users"].find_one({"Phone": phone})
    if user:
        Menuandbill(user["Name"], phone)
    else:
        print("Not found. Please register.")
elif main_choice == "3":
    phone = input("Enter Phone: ")
    order = db["orders"].find_one(
        {"userphoneno": phone}, sort=[("_id", -1)]
    )  # Latest order
    if order:
        printbill(order["order"], order["Ordervalue"])
    else:
        print("No orders found.")
```
