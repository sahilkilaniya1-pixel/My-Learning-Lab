```python
# plt.subplot(nrows, ncols, index)
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y = [10, 20, 30, 40, 50]

# Subplot 1: Line Chart
plt.subplot(1, 2, 1) # 1 row, 2 columns, index 1
plt.plot(x, y, color='blue', marker='o')
plt.title('Line Chart')

# Subplot 2: Bar Chart
plt.subplot(1, 2, 2) # Yahan '2' hona chahiye (index 2)
plt.bar(x, y, color='orange')
plt.title('Bar Chart')

# Layout ko saaf karne ke liye taaki titles aapas mein na takrayein
plt.tight_layout()
plt.show()
```

```python
import pandas as pd  # 'db' ko 'pd' kiya kyunki standard yahi hai
import matplotlib.pyplot as plt

# 1. Load the data
df = pd.read_csv('netflix_titles.csv')

# 2. Clean data
df = df.dropna(subset=['type', 'release_year', 'rating', 'country', 'duration'])

# --- Visualization 1: Bar Chart ---
type_counts = df['type'].value_counts()
plt.figure(figsize=(7, 5))
plt.bar(type_counts.index, type_counts.values, color=['skyblue', 'orange'])
plt.title('Number of Movies VS TV Shows on Netflix')
plt.xlabel('Type')
plt.ylabel('Count')
plt.savefig('movies_vs_tvshows.png')
plt.show()

# --- Visualization 2: Pie Chart ---
rating_counts = df['rating'].value_counts().head(10) # Top 10 ratings for clarity
plt.figure(figsize=(8, 6))
plt.pie(rating_counts, labels=rating_counts.index, autopct='%1.1f%%', startangle=90)
plt.title('Top 10 Content Ratings Distribution')
plt.tight_layout()
plt.savefig('content_ratings_pie.png')
plt.show()

# --- Visualization 3: Histogram (Movies Duration) ---
movie_df = df[df['type'] == 'Movie'].copy()
movie_df['duration_int'] = movie_df['duration'].str.replace(' min', '').astype(int)

plt.figure(figsize=(8, 6))
plt.hist(movie_df['duration_int'], bins=30, color='purple', edgecolor='black')
plt.title('Distribution of Movie Duration')
plt.xlabel('Duration (Minutes)')
plt.ylabel('Number of Movies')
plt.tight_layout()
plt.savefig('movies_duration_histogram.png')
plt.show()

# --- Visualization 4: Scatter Plot ---
release_counts = df['release_year'].value_counts().sort_index()
plt.figure(figsize=(10, 6))
plt.scatter(release_counts.index, release_counts.values, color='red', alpha=0.5)
plt.title('Content Release Trends Over Years')
plt.xlabel('Year')
plt.ylabel('Total Content Released')
plt.savefig('release_year_scatter.png')
plt.show()

# --- Visualization 5: Comparison Plot (Subplots) ---
content_by_year = df.groupby(['release_year', 'type']).size().unstack().fillna(0)

fig, ax = plt.subplots(1, 2, figsize=(14, 6))

# First subplot: Movies
ax[0].plot(content_by_year.index, content_by_year['Movie'], color='blue', linewidth=2)
ax[0].set_title('Movies Released Per Year')
ax[0].set_xlabel('Year')
ax[0].set_ylabel('Count')

# Second subplot: TV Shows
ax[1].plot(content_by_year.index, content_by_year['TV Show'], color='orange', linewidth=2)
ax[1].set_title('TV Shows Released Per Year')
ax[1].set_xlabel('Year')
ax[1].set_ylabel('Count')

fig.suptitle('Comparison: Movies vs TV Shows Over the Years')
plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.savefig('movies_tv_shows_comparison.png')
plt.show()
```
